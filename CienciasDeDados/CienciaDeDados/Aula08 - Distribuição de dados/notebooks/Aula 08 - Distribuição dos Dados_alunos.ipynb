# 1. Distribuição Normal dos dados

## 1.1 Gerando valores aleatórios para uma distribuição normal 

Vamos criar um conjunto de dados com uma distribuição normal onde a média das idades seja igual a 25 anos com desvio padrão igual a 5:

```python
desv_pad = 5
media = 25
```

Iremos usar a função `gauss` que gera pontos aleatórios baseados na distribuição normal. Essa função requer apenas dois atributos: Média e o desvio padrão.

```python
gauss(media, desv_pad)
```

Como temos uma função que gera valores aleatórios seguindo a distribuição normal vamos criar um laço de repetição em python que irá gerar 20000 pontos aleatórios seguindo essa distribuição:

```python
idades = []
for i in range(0, 20000) : 
    idades.append(gauss(media, desv_pad))
```

Vamos checar os tipos dos dados que criamos:

```python
type(idades)
idades
```

Poderiamos usar uma lista para plotar gráficos dos nossos dados. Entretanto, iremos convertê-lo para um DataFrame do pandas para podermos usar todas as outras ferramentas que essa biblioteca nos disponibiliza.

```python
df_idades = pd.DataFrame(idades, columns=['idades'])
df_idades
```

```python
media_real = df_idades['idades'].mean()
media_real
```

```python
h = np.histogram(df_idades['idades'], bins=20)
h
```

```python
sns.histplot(df_idades['idades'], bins=20, kde= True)
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.axvline(x=media_real, color='red', linestyle = 'dashed', linewidth = 2)
```
